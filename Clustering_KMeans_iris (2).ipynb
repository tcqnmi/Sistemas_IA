```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

datos = datasets.load_iris()
#x = StandardScaler().fit_transform(datos.data)
x = datos.data
x
```

```python
km = KMeans(n_clusters=3,  random_state=24)
modelo=km.fit(x)
y_pred = km.fit_predict(x)
```

```python
print("Agrupaciones: (K-means)\n", km.labels_)
```

```python
print("Observaciones etiquetadas en el dataset original:\n", datos.target)
```

```python
print(y_pred)
```

```python
# salida grafica

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x)

km_pca = KMeans(n_clusters=3, random_state=24)
km_pca.fit(x_pca)
y_km = km_pca.fit_predict(x_pca)

import matplotlib.pyplot as plt
plt.scatter(x_pca[:,0], x_pca[:,1], c=y_km, cmap='viridis', edgecolors="k", s=50)
centers = km_pca.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='red', marker='x', s=200, alpha=0.75)
plt.title("Agrupamiento con K-means")
plt.xlabel('Eje X1')
plt.ylabel('Eje X2')
plt.show()
```

```python
from sklearn.metrics import homogeneity_score, completeness_score, v_measure_score

print(f"Homogeneidad: {homogeneity_score(datos.target, modelo.labels_):.3f}")
print(f"Plenitud: {completeness_score(datos.target, modelo.labels_):.3f}")
print(f"V-measure: {v_measure_score(datos.target, modelo.labels_):.3f}")
```

```python
#Cálculo de la pureza 
from sklearn.metrics import confusion_matrix
import numpy as np

cm = confusion_matrix(datos.target, y_pred)

#Suma total de los máximos por fila dividido entre la suma total de muestras
pureza = np.sum(np.amax(cm, axis= 0))/np.sum(cm)

print("Pureza: ", pureza)
```

```python
from sklearn.metrics import silhouette_score

y = datos['target']
res = silhouette_score(x,y)
print(f"Coeficiente de Silhouette medio de todas las muestras: {res:.2f}")

#Coeficiente Silhouette para cada una de las muestras
from sklearn.metrics import silhouette_samples
import pandas as pd

detalle = silhouette_samples(x,y)
df = pd.DataFrame(x, columns = datos.feature_names)
df['cluster']=modelo.labels_
df['sil']=detalle
df
```

```python
inertia = []

#Probamos diferente valores de k sobre el modelo 
for k in range(1,11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)

#Varianza (en función del valor de k)
plt.plot(range(1,11), inertia, marker = 'o')
plt.title("Método del Codo")
plt.xlabel("Número de clusters (k)")
plt.ylabel('Varianza')
plt.show()    

#Con este "codo" lo que buscamos es encontrar ese punto en el que la agrupación sea más significativa, 
#hasta ahora usábamos el 3 porque sabíamos que teníamos 3 variedades de flores
```
