```python
from sklearn.datasets import make_blobs

x, y, centros = make_blobs(
    n_samples=100,
    n_features=2,
    centers=3,
    cluster_std=1.8,
    random_state=24,
    return_centers=True
)

import matplotlib.pyplot as plt

plt.scatter(x[:,0], x[:,1], c=y, cmap='viridis', s=50)
plt.scatter(centros[:,0], centros[:,1], c = 'red', marker='X')
plt.title("Nubes de puntos (antes)")
plt.xlabel('Eje X1')
plt.ylabel('Eje X2')
plt.show()
```

```python
from sklearn.cluster import KMeans
# Aplicamos K-means con k=3 (número de grupos)
kmeans = KMeans(n_clusters=4, n_init="auto")
kmeans.fit(x)
```

```python
# Obtenemos los centroides y las etiquetas asignadas a los grupos
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Representación gráfica
plt.scatter(x[:,0], x[:,1], c = labels, cmap='viridis', s=50)
plt.scatter(
    centroids[:,0],
    centroids[:,1],
    color = 'red',
    marker= 'x',
    label='Centroides',
    s=50
)
plt.title("Agrupamiento con K-means")
plt.xlabel('Eje X1')
plt.ylabel('Eje X2')
plt.legend()
plt.show()
```
